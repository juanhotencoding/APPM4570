## Problem 1

The manager of a web company is interested in predicting the number of megabytes downloaded by clients (*i.e.* the variable ${\tt megasd}$) according to the number of minutes they are connected (*i.e.* the variable ${\tt mconnected}$). The manager randomly selects clients and records the data (*i.e.* the (${\tt megasd, mconnected}$) pairs) and stores them in the file ${\tt url-SP19-R.csv}$ (or ${\tt url-SP19-TAB.txt}$).

#### (a) Load the data from ${\tt url-SP19-R.csv}$ (or ${\tt url-SP19-TAB.txt}$).

```r
web_data <- read.csv("url-SP19-R.csv")
head(web_data)
```

#### (b) Calculate the covariance and correlation between the two sets of data, *i.e.* between ${\tt megasd}$ and ${\tt mconnected}$.


```r
cov.megasd.mconnected <- cov(web_data$megasd, web_data$mconnected)
cat("The covariance between megasd and mconnected is", cov.megasd.mconnected)
```

```r
corr.megasd.mconnected <- cor(web_data$megasd, web_data$mconnected)
cat("The correlation between megasd and mconnected is", corr.megasd.mconnected)
```

#### (c) We showed in class that if $\rho(X,Y)=1$ then 
$$
\displaystyle Y=a+bX=-\sigma_{Y}\left(\dfrac{\mu_{X}}{\sigma_{X}}
+\dfrac{\mu_{Y}}{\sigma_{Y}}\right)
+\dfrac{\sigma_{Y}}{\sigma_{X}}X.
$$

#### Now if we let $X$ be the number of minutes a client is connected (${\tt mconnected}$) and $Y$ be the megabytes downloaded (${\tt megasd}$), calculate $a$ and $b$ above and display their value.

```r
mu.x <- mean(web_data$mconnected)
sigma.x <- sd(web_data$mconnected)
mu.y <- mean(web_data$megasd)
sigma.y <- sd(web_data$megasd)

a <- -sigma.y*((mu.x/sigma.x) + (mu.y/sigma.y))
b <- sigma.y / sigma.x

cat("a is ", a, " b is", b)
```

#### (d) Create a plot of the data with ${\tt mconnected}$ on the x-axis and ${\tt megasd}$ on the y-axis. Do the variables look correlated?

```r
plot(web_data$mconnected, web_data$megasd, xlab="number of minutes a client is connected", ylab="megabytes downloaded")
```

**STUDENT ANSWER**: The variables do look to be positively and linearly correlated. The more time the client is connected the amount of downloads (in megabytes) increases.

## Problem 2

#### (a) Generate three random vectors of length $n = 500$ from:

1. $X \sim beta(3,1)$
2. $Y \sim beta(3,1)$
3. $Z \sim beta(2,3)$

#### (see the ${\tt rbeta()}$ command) and then use the ${\tt head()}$ command to preview the vectors you generated. 

```r
set.seed(12)
n = 500
rv.X <- rbeta(n = n, shape1 = 3, shape2 = 1)
cat("Random Variable X: ", head(rv.X))
rv.Y <- rbeta(n = n, shape1 = 3, shape2 = 1)
cat("\nRandom Variable Y: ", head(rv.Y))
rv.Z <- rbeta(n = n, shape1 = 2, shape2 = 3)
cat("\nRandom Variable Z: ", head(rv.Z))
```

#### (b) Create a plot of $X$ and $Y$. Do the variables look correlated? What is $\rho(X,Y)$?

```r
plot(rv.X, rv.Y)
```

**STUDENT ANSWER:** The variables do not look like they are correlated at all. However, even though they do not look to be correlated in a linear manner we can not rule out the possibility of these variables being correlated in a non-linear manner.

```r
cat("The correlation value of the random variables is:", cor(rv.X, rv.Y), "\nwhich is very close to zero, telling us that these variables are in fact are not correlated.")
```

#### (c) Create vectors ${\tt xstar}$ and ${\tt ystar}$ by taking $X$ and $Y$ and dividing by $Z$.  Are ${\tt xstar}$ and ${\tt ystar}$ correlated? Support your claim with numerical and graphical evidence similar to what was done in Problem 1(e). 

```r
xstar <- rv.X / rv.Z

ystar <- rv.Y / rv.Z

plot(xstar, ystar)
```

```r
cat("The correlation value of the random variables xstar and ystart equals:", cor(xstar, ystar))
cat("\nWhich is very close to 1 so we can say that the random variables are correlated linear and positively")
```

```r
mu.xstar <- mean(xstar)
sigma.xstar <- sd(xstar)
mu.ystar <- mean(ystar)
sigma.ystar <- sd(ystar)

a <- -sigma.ystar*((mu.xstar/sigma.xstar) + (mu.ystar/sigma.ystar))
b <- sigma.ystar / sigma.xstar

cat("a is ", a, " b is", b)
```

```r
plot(xstar, ystar, main = "Line of best fit")
par(new=T)
plot(xstar, a+(xstar*b), type="l", xlab='', ylab='',xaxt='n', yaxt='n',col='red')
legend("top", legend=TeX('$Y=-\\sigma_{Y}\\left(\\frac{\\mu_{X}}{\\sigma_{X}}+\\frac{\\mu_{Y}}{\\sigma_{Y}}\\right)+X\\frac{\\sigma_{Y}}{\\sigma_{X}}
$'),
       col="red", lty=1:2, cex=0.8)
```

#### (d) Interpret and analyze the results from the previous part. What practical lessons can you learn from this example?

**STUDENT ANSWER:** The partical lessons that I learned from these examples are that if you are looking to see if two random variables are correlated or not, it can be useful to plot them on the x-y plane. This will give you the overall shape of the data. You can then fit a line over that data to see if they are linearly dependent. Also calculating the correlation value rho is useful to check against the graph. If the rho value is close to -1 or 1 then you can say the random variables are correlated if the rho value is close to zero then you can say they are not correlated. This is of course limited to the 2-D case and it will be harder to visualize and plot higher dimensional data.
