## Problem 1

There is an urn that has 100 marbles in it: 2 are red, 70 are yellow, 13 are green and 15 are blue. In this problem you will simulate data to imitate someone drawing a marble from the urn (replacing after each draw) 1000 times, with the color recorded at each draw (HINT: the ${\tt sample()}$ function in ${\tt R}$ will be useful here).


#### (a) Construct the urn, that is, construct a vector that has 100 entries: 2 red ("r"), 70 yellow ("y"), 13 green ("g").

```r
urn <- c(rep("r", times=2), rep("y", times=70), rep("g", times=13), rep("b", times=15))
```

```r
urn
```

#### (b) Simulate drawing a marble from the urn recording it's color and putting it back $1000$ times, use ${\tt sample()}$.

```r
set.seed(999)
n = 1000
drawing <- sample(urn, size=n, replace = TRUE)
```

```r
head(drawing)
```

```r
length(drawing)
```

#### (c) Plot the fraction of times a yellow marble is drawn in the first 1000 draws. On the x-axis, include the number of draws, and on the y-axis, include the cumulative fraction of yellow draws. Why does this number fluctuate more when the number of draws is smaller?

```r
frac.y <- cumsum(urn == "y")/(1:n)
```

```r
plot(frac.y, type="l", xlab="Number of draws", ylab="Fraction of Yellow", col = "yellow")

ratio.y <- frac.y[n]
abline(ratio.y, 0)

cat("The ratio of yellow draws is", ratio.y, "\n")
```

**STUDENT ANSWER:** The number fluctuates because the cumulative sum is divided by the current number of draws and it is also the case that yellow has a high probability of being drawed because that is the color we have the most of, but, because we are drawing randomly there are still draws in which yellow is not drawn and so on that case we would assign a zero probability. 

#### (d) Repeat part 1(a)-(c) above for the red marble. What do you notice that is different, and why do you think it is different?



```r
frac.r = cumsum(urn == "r")/(1:n)

plot(frac.r, type="l", xlab="Number of draws", ylab="Fraction of Red", col = "red")

ratio.r <- frac.r[n]
abline(ratio.r, 0)

cat("The ratio of red draws is", ratio.r, "\n")
```

**STUDENT ANSWER** In the case of drawing the color red, the plot shows a more smooth progression to the overall probability of reds we have, namely, $2\%$. This is because we observe red draws less frequently as opposed to observing draws of the color yellow.
